Visualizations

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.boxplot(x='host_tier', y='charges?', data = df_hospital)
plt.title('Distribution of Charge across hospital tiers')
plt.xlabel('Hospital Tier')
plt.ylabel('Charges')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='yr', y='charges?', data=df_hospital)
plt.title('Yearly Trend of Charges')
plt.xlabel('Year')
plt.ylabel('Charges')
plt.show()
```

```python
plt.figure(figsize=(10,8))
correlation_matrix = df_hospital.corr()
sns.heatmap(correlation_matrix, annot = True, cmap ='coolwarm', fmt = '2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numerical_vars = ['b_m_i', 'HBA1C', 'noofmajorsurgeries', 'recovery_period']
sns.pairplot(df_exam[numerical_vars])
plt.suptitle('Pairplot of Numerical Varialble',y=1.02)
plt.show()
```

```python
correlation_matrix = df_exam[['b_m_i', 'HBA1C', 'noofmajorsurgeries','recovery_period']].corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot = True, cmap='coolwarm', fmt ='.2f')
plt.title('Correlation Heatmap for Numerical Variables')
plt.show()
```

```python
plt.figure(figsize=(12,6))
numerical_vars = ['b_m_i', 'HBA1C', 'noofmajorsurgeries', 'recovery_period' ]
sns.boxplot(data = df_exam[numerical_vars])
plt.title('Boxplot for Numerical Variables')
plt.show()
```
